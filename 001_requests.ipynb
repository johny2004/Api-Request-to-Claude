```python
# Load env variables 
from dotenv import load_dotenv

load_dotenv()
```

```python
#Create an API Client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0" 
```

```python
#Make a request
message = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=[
        {
            "role": "user",
            "content": "What is quantum computing? Answer in one sentence"
        }
    ]
)
```

```python
#Catch the answer
message # this brings a lot of information more than just only the content
message.completion # this is the content of the answer
message.content[0].text # this is the content of the answer, but in a different way to access it. You can use either one.
```
