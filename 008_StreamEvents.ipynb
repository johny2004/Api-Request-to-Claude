```python
#Create an API Client
from anthropic import Anthropic

client = Anthropic()
model = "claude-sonnet-4-0" 
```

```python
#MultiChat Helpers

def add_user_message(messages, text):
    user_message = {"role": "user", "content": text}
    messages.append(user_message)

def add_assistant_message(messages, text):
    assistant_message = {"role": "assistant", "content": text}
    messages.append(assistant_message)
    
def chat(messages, system=None):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages
    }
    
    if system:
        params["system"] = system
    
    message = client.messages.create(**params)
    return message.content[0].text
```

```python
#Basic Stream Implementation
    #It calls back all the severyl types of events, not only the content of the answer, but also when the message starts, when it ends, when there is an error, etc.
messages = []
add_user_message(messages, "Write a 1 sentence description of a fake database")

stream = client.messages.create(
    model=model,
    max_tokens=1000,
    messages=messages,
    stream=True
)

for event in stream:
    print(event)
```

```python
#Simplified Text Streaming
# extracts just the text content:
#Not just get back one word of each event but a chunk of text, which is more useful when you want to display the answer as it's being generated.
messages = []
add_user_message(messages, "Write a 1 sentence description of a fake database")


with client.messages.stream(
    model=model,
    max_tokens=1000,
    messages=messages
) as stream:
    for text in stream.text_stream:
        print(text, end="")
```
